```python
from transformers import pipeline

# Load the model pipeline
summarizer = pipeline(task="summarization", model="cnicu/t5-small-booksum")

# Long text about the Eiffel Tower
long_text = """The Eiffel Tower is a wrought-iron lattice tower on the Champ de Mars in Paris, France.
It is named after the engineer Gustave Eiffel, whose company designed and built the tower.
Constructed from 1887 to 1889 as the entrance arch for the 1889 World's Fair,
it was initially criticized by some of France's leading artists and intellectuals for its design,
but it has become a global cultural icon of France and one of the most recognizable structures in the world.
The tower is the most-visited paid monument in the world, with millions of people ascending it each year.
It was the tallest man-made structure in the world until the completion of the Chrysler Building in New York in 1930.
The tower has three levels for visitors, with restaurants on the first and second levels.
The top level's upper platform is 276 meters (906 feet) above the ground, the highest observation deck
accessible to the public in the European Union."""

# Pass the long text to the model
output = summarizer(long_text, max_length=50, min_length=25, do_sample=False)

# Access and print the summarized text
print(output[0]['summary_text'])
```

```python
# Instantiate the pipeline
generator = pipeline(task="text-generation", model="gpt2")

response = "Dear valued customer, I am glad to hear you had a good stay with us."
text = """The Eiffel Tower is a wrought-iron lattice tower on the Champ de Mars in Paris, France.
It is named after the engineer Gustave Eiffel, whose company designed and built the tower.
Constructed from 1887 to 1889 as the entrance arch for the 1889 World's Fair,
it was initially criticized by some of France's leading artists and intellectuals for its design,
but it has become a global cultural icon of France and one of the most recognizable structures in the world.
The tower is the most-visited paid monument in the world, with millions of people ascending it each year.
It was the tallest man-made structure in the world until the completion of the Chrysler Building in New York in 1930.
The tower has three levels for visitors, with restaurants on the first and second levels.
The top level's upper platform is 276 meters (906 feet) above the ground, the highest observation deck
accessible to the public in the European Union."""

# Pass the long text to the model

# Complete the prompt
prompt = f"Customer review:\n{text}\n\nHotel reponse to the customer:\n{response}"

# Complete the model pipeline
# Use max_new_tokens to control the length of generated text
outputs = generator(text, max_new_tokens=50, pad_token_id=generator.tokenizer.eos_token_id, truncation=True)

print(outputs[0]["generated_text"])
```

```python
from transformers import pipeline

# Spanish text to translate
spanish_text = "Este curso sobre LLMs se está poniendo muy interesante"

# Define the pipeline
translator = pipeline(task="translation_es_to_en", model="Helsinki-NLP/opus-mt-es-en")

# Translate the Spanish text
translations = translator(spanish_text, clean_up_tokenization_spaces=True)

print(translations[0]["translation_text"])
```
